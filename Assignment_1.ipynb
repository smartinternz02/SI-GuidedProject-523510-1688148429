```python
import pandas as pd
import numpy as np
```

**Task-1**

```python
#create dataframe
df = pd.DataFrame([np.random.rand(4), np.random.rand(4), np.random.rand(4), np.random.rand(4)])
df
```

**Task-2**

```python
#Renaming the columns of the dataframe
df.rename(columns={0: 'Random value 1', 1: 'Random value 2', 2: 'Random value 3', 3: 'Random value 4'}, inplace=True)
df
```

**Task-3**

```python
df.describe()
```

**Task-4**

```python
#Finding null values
print(df.isnull(), '\n')

#Displaying the datatype of each column
print('Data type of Random value 1: {}'.format(df['Random value 1'].dtype))
print('Data type of Random value 2: {}'.format(df['Random value 2'].dtype))
print('Data type of Random value 3: {}'.format(df['Random value 3'].dtype))
print('Data type of Random value 4: {}'.format(df['Random value 4'].dtype))
```

**Task-5**

```python
#location method
print(df.loc[0:,'Random value 2':'Random value 3'], '\n')

#Index location method
print(df.iloc[0:, 1:3])
```
